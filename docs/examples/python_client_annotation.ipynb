# Ranking annotations from METASPACE python client data

In this workflow we will download METASPACE data using the pythoon client and rank the annotations with the package.

```python
import metaspace
import linex2metaspace as lx2m
import pandas as pd
```

We will download a dataset from METASPACE: `2016-09-21_16h06m56s` (Selected because of a good number of annotations).

```python
sm = metaspace.SMInstance()
ds = sm.dataset(id='2016-09-21_16h06m56s')

# Downloading the annotations
annotations = ds.results(database=('HMDB', 'v4'), fdr=.1)
```
